```python
import pandas as pd # библиотека для работы с данными
import fastai # библиотека для работы с нейронными сетями
from fastai.tabular.all import* # блок для связки нейронных сетей и табличных данных
train_data = pd.read_csv("./drive/MyDrive/spaceship-titanic/train.csv")
train_data = train_data.drop("PassengerId", axis = 1)
train_data.head()
```

```python
train_data.dtypes
```

```python
c=[]
n=[]
for i in train_data.columns:
    if(train_data[i].dtype) in ['object']:
        c.append(i)
stay=["Transported"]

for i in train_data.columns:
    if i not in stay and i not in c:
        n.append(i)

train_data[c].info()
```

```python
train_data[n].info()
```

```python
len(c), len(n)
```

```python
train_data.info()
```

```python
import fastai
from fastai.tabular.all import *
import torch

splits = RandomSplitter(valid_pct = 0.2)(range_of(train_data))
to = TabularPandas(train_data, procs = [Normalize, Categorify, FillMissing],
                   cat_names = c,
                   cont_names = n,
                   y_names = 'Transported',
                  #  y_block = RegressionBlock,
                   splits = splits)

dls = to.dataloaders(bs = 128)
dls.show_batch()
```

```python
learn = tabular_learner(dls,metrics=accuracy,lr=0.05)
learn.fit_one_cycle(100)
```

```python
learn.show_results()
```
