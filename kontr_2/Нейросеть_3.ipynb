```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
import fastai
from fastai.tabular.all import *
!pip install kaggle
!pip install opendatasets
import opendatasets as od
od.download("https://www.kaggle.com/datasets/uciml/iris")
train = pd.read_csv("/content/iris/Iris.csv")
```

```python
train.info()
```

```python
train = train.drop("Id", axis = 1)
train.head()
```

```python
train.dtypes
```

```python
n=[]
stay=["Species"]
for i in train.columns:
    if i not in stay:
        n.append(i)
train[n].info()
```

```python
import torch
splits = RandomSplitter(valid_pct = 0.2)(range_of(train))
to = TabularPandas(train, procs = [Normalize, Categorify, FillMissing],
                  #  cat_names = c,
                   cont_names = n,
                   y_names = 'Species',
                  #  y_block = RegressionBlock,
                   splits = splits)

dls = to.dataloaders(bs = 64)
dls.show_batch()
```

```python
learn = tabular_learner(dls,metrics=accuracy,lr=0.05)
learn.fit_one_cycle(50)
```

```python
learn.show_results()
```
